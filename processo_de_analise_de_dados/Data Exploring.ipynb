```python
import pandas as pd

%matplotlib inline     
# Permite que os gráficos sejam visializado em células do jupyter

df = pd.read_csv('df_data_set_claer.csv')
df.info()
```

### Histograma

```python
df.hist(figsize=(8,8));  # ';' faz com que outputs indesejados sejam exibidos...
```

### Criando histogramas para variáveis específicas...

O histograma, também conhecido como distribuição de frequências, é a representação gráfica em colunas ou em barras (retângulos) de um conjunto de dados previamente tabulado e dividido em classes uniformes ou não uniformes.A base de cada retângulo representa uma classe. A altura de cada retângulo representa a quantidade ou a frequência absoluta com que o valor da classe ocorre no conjunto de dados para classes uniformes ou a densidade de frequência para classes não uniformes.

```python
### Podemos criar histogramas para dados específicos:
df['age'].hist();
```

### Podemos usar também a função geral de plotagem de gráficos:

```python
df['age'].plot(kind='hist');
```

### Gráfico de Barras

Para criar um gŕafico de barras precisamos de valores de contagem, um para cada categoria que será apresentada.
Para isso usamos a função `value_counts()`.

```python
df['education'].value_counts()
```

#### A partir dessa função podemos gerar o gráfico de barras:

```python
df['education'].value_counts().plot(kind='bar')# se usar ';' no final podemos retirar o output desnecessário abaixo
```

## Gráficos de Pizza

Para usarmos gráficos de pizza também precisamos de `value_counts()`.

```python
df['workclass'].value_counts().plot(kind='pie', figsize=(8,8));
```

```python
cancer = pd.read_csv('cancer_means.csv')
cancer.head()
```

```python
cancer.info()
```

A próxima função é muito legal pois permite um `insight` rápido na relação entre variáveis numéricas com gráficos de dispersão(scatter).

```python
pd.plotting.scatter_matrix(cancer, figsize=(15,15));
```

#### Podemos criar um único gráfico de dispersão usando a função de gráficos com parâmetros para especificar as colunas que serão usadas nos eixos X e Y.

```python
cancer.plot(x='compactness_mean', y='concavity_mean', kind='scatter');
```

```python
# E finalmente podemos criar um gráfico de caixa
cancer['concave_points_mean'].plot(kind='box');
```
